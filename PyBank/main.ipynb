```python
# Import the pathlib and csv library
from pathlib import Path
import csv

# Set the file path
csvpath = Path("budget_data.csv")


# Initialize the metric variables
count = 0
total = 0
average_change = 0
minimum = 0
maximum = 0
month_pnl=0

# Open the csv file as an object
with open(csvpath, "r") as csvfile:

    # Pass in the csv file to the csv.reader() function
    # (with ',' as the delmiter/separator) and return the csvreader object
    csvreader = csv.reader(csvfile, delimiter=",")

    # Read the header row first (skip this step if there is no header)
    csv_header = next(csvreader)


 # Read each row of data after the header
    for row in csvreader:


        # Set the 'date' and 'pnl' variables for better
        # readability, convert strings to ints for numerical calculations
        date = row[0]
        pnl = int(row[1])

    # Cumulatively sum up the total profits/losses and count of trading months
        total += pnl
        count += 1

    # Logic to determine minimum and maximum values
        if minimum == 0:
            minimum = pnl
            minimum_date = date
        elif pnl < minimum:
            minimum = pnl
            minimum_date =date
        elif pnl > maximum:
            maximum = pnl
            maximum_date = date

# Calculate the average, round to the nearest 2 decimal places   
average_change = round(total/count,2)

# Print all the calculated results
print('Financial Analysis')
print(f'---------------------------------------------------')
print(f'Total months: {count}')
print(f'Total: {total}')
print(f'Average Change: $ {average_change}')
print(f'Greatest Increase in Profits: {maximum_date} ($ {maximum})')
print(f'Greatest Decrease in Profits: {minimum_date} ($ {minimum})')
print(f'---------------------------------------------------')

# Set output file name
output_path = 'output.txt'

# Open the output path as a file object
with open(output_path, 'w') as file:
    # Write all the results to the output file
    file.write(f'Financial Analysis\n')
    file.write(f'---------------------------------------------------\n')
    file.write(f'Total months: {count}\n')
    file.write(f'Total: {total}\n')
    file.write(f'Average Change: $ {average_change}\n')
    file.write(f'Greatest Increase in Profits: {maximum_date} ($ {maximum})\n')
    file.write(f'Greatest Decrease in Profits: {minimum_date} ($ {minimum})\n')
    file.write(f'---------------------------------------------------\n')
```
